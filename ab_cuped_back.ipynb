```python
import pandas as pd
import numpy as np
#убрать топ 10 ковариат
pd.set_option('display.max_columns', None)

file_name = 'dataset_mb (1).csv.gz'

print("Загружаем данные...")
df_mb = pd.read_csv(file_name, compression='gzip')

print(f"Размер исходного датасета: {df_mb.shape}")

df_mb = df_mb.replace(-1, 0)

print(df_mb.isnull().sum()[df_mb.isnull().sum() > 0])

df_mb = df_mb.fillna(0)

print(f"\nОсталось Null значений: {df_mb.isnull().sum().sum()}")

display(df_mb.head())
```

```python
print("\n--- 1. Проверка распределения целевых метрик (Таргетов) ---")
targets = ['life_1_day_flg', 'life_32_day_flg', 'orig_cost_sum']

for col in targets:
    if col in df_mb.columns:
        print(f"\nСтатистика по метрике '{col}':")
        if 'flg' in col:
            print("Распределение (в %):")
            print((df_mb[col].value_counts(normalize=True) * 100).round(2))
        else:
            print(df_mb[col].describe().round(2))
    else:
        print(f"ВНИМАНИЕ: Колонка '{col}' не найдена в датасете!")

print("\n--- 2. Проверка главных кандидатов в ковариаты ---")
scores = ['attririon_score', 'loyalty_score']
for col in scores:
    if col in df_mb.columns:
        print(f"{col:<17} | Min: {df_mb[col].min():.2f} | Max: {df_mb[col].max():.2f} | Среднее: {df_mb[col].mean():.2f}")

print("\nДанные проверены! Готовы к расчету CUPED.")
```

```python
print("Проверяем гипотезу: Бинаризация attrition_score по порогу > 0.3\n")

target = 'life_1_day_flg'


df_mb['attrition_score_bin'] = (df_mb['attrition_score'] > 0.3).astype(int)

corr_original = df_mb[target].corr(df_mb['attrition_score'])
corr_bin = df_mb[target].corr(df_mb['attrition_score_bin'])

print(f"Связь с {target}:")
print(f"Оригинальный attrition_score (непрерывный): {abs(corr_original):.4f}")
print(f"Бинаризованный attrition_score_bin (>0.3):  {abs(corr_bin):.4f}")

if abs(corr_bin) > abs(corr_original):
    print("\n🔥 Успех! Бинаризация сработала: мы убрали шум и усилили сигнал.")
    print("Теперь матрица корреляций на Шаге 3 автоматически подхватит новую фичу!")
else:
    print("\n📉 Связь упала. При обрезке порога мы потеряли ценную информацию (дисперсию).")
    print("Лучше использовать оригинальный скор для линейного CUPED.")
```

```python
import scipy.stats as stats

# Шаг 4: CUPED и расчет экономии выборки
target = 'life_1_day_flg'

covariate = 'attrition_score'


covariance = df_mb[target].cov(df_mb[covariate])
variance_x = df_mb[covariate].var()
theta = covariance / variance_x

mean_x = df_mb[covariate].mean()
df_mb[f'{target}_cuped'] = df_mb[target] - theta * (df_mb[covariate] - mean_x)

var_original = df_mb[target].var()
var_cuped = df_mb[f'{target}_cuped'].var()
variance_reduction = (1 - var_cuped / var_original) * 100

# var(Y_cuped) = var(Y) * (1 - rho^2)
rho = df_mb[target].corr(df_mb[covariate])
theoretical_reduction = (rho**2) * 100


alpha = 0.05       # Ошибка 1 рода (5%)
power = 0.80       # Мощность (80%) -> Ошибка 2 рода (Beta = 0.20)
mde_absolute = 0.01 # Ловим абсолютный эффект в +1%


z_alpha = stats.norm.ppf(1 - alpha / 2)
z_beta = stats.norm.ppf(power)
z_factor = (z_alpha + z_beta)**2

# Расчет выборки на 1 группу
n_original = (2 * var_original * z_factor) / (mde_absolute**2)
n_cuped = (2 * var_cuped * z_factor) / (mde_absolute**2)

sample_size_reduction = (1 - n_cuped / n_original) * 100


print(f"Коэффициент Тета (θ): {theta:.6f}")
print(f"Корреляция (ρ):       {rho:.4f}")

print("\n--- РЕЗУЛЬТАТЫ CUPED ---")
print(f"Оригинальная дисперсия:   {var_original:.6f}")
print(f"Новая дисперсия CUPED:    {var_cuped:.6f}")
print(f"🔥 Теоретическое сжатие (ρ^2): {theoretical_reduction:.2f}%")
print(f"🔥 Фактическое сжатие:         {variance_reduction:.2f}%\n")

print("--- ЭКОНОМИЯ ТРАФИКА В А/Б ТЕСТЕ ---")
print(f"Параметры теста: MDE = {mde_absolute*100}%, Alpha = {alpha}, Power = {power}")
print(f"Требуемый трафик ДО CUPED (на 1 группу):    {int(n_original):,} пользователей")
print(f"Требуемый трафик ПОСЛЕ CUPED (на 1 группу): {int(n_cuped):,} пользователей")
print(f"🔥 Выборка сократилась на:                  {sample_size_reduction:.2f}%")
print(f"Чистая экономия трафика:                    {int(n_original - n_cuped):,} пользователей!")
```

```python
import numpy as np
import pandas as pd

target = 'life_1_day_flg'

# 1. СТРОГИЙ СПИСОК ИСКЛЮЧЕНИЙ
excluded_columns = [
    'life_1_day_flg', 'life_32_day_flg', 'orig_cost_sum',
    'nfu_flg', 'channel_desc', 'flow_strategy_code',
    'app_version_no', 'app_nm', 'full_apps'
]

# Выбираем только числовые фичи
numeric_df = df_mb.select_dtypes(include=[np.number])

# Отсекаем все метрики, которые мы насоздавали на прошлых шагах!
features = [col for col in numeric_df.columns
            if col not in excluded_columns and not col.endswith('_cuped')]

print(f"Запускаем Итеративный CUPED. Доступно независимых признаков: {len(features)}\n")


y_iter = df_mb[target].copy()
var_original = y_iter.var()

used_features = []
max_iterations = 1000

for i in range(max_iterations):
    # Считаем корреляции с ТЕКУЩИМ остатком (Y_step1_all, Y_step2_all и т.д.)
    corrs = df_mb[features].corrwith(y_iter).abs()
    corrs = corrs.drop(labels=used_features, errors='ignore')

    if corrs.empty or corrs.max() < 0.005: # Если связи слишком слабые - останавливаемся
        print("Оставшиеся фичи слишком слабые. Остановка.")
        break

    best_feature = corrs.idxmax()
    best_corr = corrs.max()
    used_features.append(best_feature)

    # === Шаг расчета Тета (Аналог: theta1 = np.cov(Y, X) / np.var(X)) ===
    cov = y_iter.cov(df_mb[best_feature])
    var_x = df_mb[best_feature].var()

    if var_x == 0: continue

    theta = cov / var_x
    mean_x = df_mb[best_feature].mean()

    # === Обновление остатков (Аналог: Y_step1 = Y - theta * (X - X_mean)) ===
    y_iter = y_iter - theta * (df_mb[best_feature] - mean_x)

    variance_reduction = (1 - y_iter.var() / var_original) * 100
    print(f"Итерация {i+1}: Фича = '{best_feature:<17}' | Корр. с остатком = {best_corr:.4f} | Общее сжатие = {variance_reduction:.2f}%")

# Сохраняем финальный результат
df_mb[f'{target}_multi_cuped'] = y_iter

var_single_cuped = df_mb[f'{target}_cuped'].var()
var_multi_cuped = df_mb[f'{target}_multi_cuped'].var()

print("\n--- ИТОГОВОЕ СРАВНЕНИЕ ДИСПЕРСИЙ ---")
print(f"1. Оригинальная дисперсия:      {var_original:.6f}")
print(f"2. Дисперсия Single-CUPED:      {var_single_cuped:.6f}")
print(f"3. Дисперсия Iterative-CUPED:   {var_multi_cuped:.6f}")
print(f"🔥 Итоговое сжатие Iterative:   {(1 - var_multi_cuped / var_original) * 100:.2f}%")
```

```python
import numpy as np
import pandas as pd
import scipy.stats as stats
from sklearn.ensemble import HistGradientBoostingRegressor
from sklearn.model_selection import KFold, cross_val_predict

print("--- Шаг 6: CUPAC ---")

target = 'life_1_day_flg'


# Указываем, какие колонки ТОЧНО нельзя брать в модель
bad_columns = [
    'life_1_day_flg', 'life_32_day_flg', 'orig_cost_sum',
    'nfu_flg', 'channel_desc', 'flow_strategy_code',
    'app_version_no', 'app_nm', 'full_apps'
]

# Отбираем числовые колонки, вычищая всё, что мы создали на прошлых шагах!
numeric_cols = df_mb.select_dtypes(include=[np.number]).columns
features = [
    col for col in numeric_cols
    if col not in bad_columns
    and 'cuped' not in col.lower()  # Исключаем все прошлые cuped
    and 'cupac' not in col.lower()  # Исключаем прошлые cupac
    and 'pred' not in col.lower()   # Исключаем прошлые предсказания
    and 'bin' not in col.lower()    # Исключаем бинаризованные эксперименты
]

print(f"Для обучения ML отобрано {len(features)} чистых признаков")

X = df_mb[features]
Y = df_mb[target]

print("\n⏳ Обучаем ML-модель (HistGradientBoosting) и получаем OOF прогнозы...")

# Используем KFold с перемешиванием для честной кросс-валидации
kf = KFold(n_splits=5, shuffle=True, random_state=42)
ml_model = HistGradientBoostingRegressor(max_iter=100, max_depth=4, random_state=42)

# Получаем честные прогнозы (модель не видит таргет тех строк, которые предсказывает)
oof_predictions = cross_val_predict(ml_model, X, Y, cv=kf)

# Сохраняем как новую "супер-ковариату"
df_mb['ml_covariate'] = oof_predictions

# Считаем ковариацию между таргетом и ML-прогнозом
covariance = df_mb[target].cov(df_mb['ml_covariate'])
variance_pred = df_mb['ml_covariate'].var()

# Вычисляем коэффициент Тета (с защитой от деления на ноль)
theta_cupac = covariance / variance_pred if variance_pred > 0 else 0
mean_pred = df_mb['ml_covariate'].mean()

# Формула из статьи: Y_cupac = Y - theta * (Y_pred - E[Y_pred])
df_mb[f'{target}_cupac'] = df_mb[target] - theta_cupac * (df_mb['ml_covariate'] - mean_pred)

# === 4. ЧЕСТНАЯ ОЦЕНКА ДИСПЕРСИИ ===
var_original = df_mb[target].var()
var_cupac = df_mb[f'{target}_cupac'].var()
reduction_cupac = (1 - var_cupac / var_original) * 100

print("\n--- РЕЗУЛЬТАТЫ СЖАТИЯ ДИСПЕРСИИ ---")
print(f"Оригинальная дисперсия:       {var_original:.6f}")
print(f"Дисперсия после CUPAC:        {var_cupac:.6f}")
print(f"🔥 Честное сжатие дисперсии:   {reduction_cupac:.2f}%")
print(f"Коэффициент Тета (θ):         {theta_cupac:.4f}")

#теоретическое сжатие должно быть примерно равно R^2 (квадрату корреляции)
rho = df_mb[target].corr(df_mb['ml_covariate'])
print(f"Корреляция Таргета и Прогноза (ρ): {rho:.4f}")
print(f"Теоретическое сжатие (ρ^2):   {(rho**2) * 100:.2f}%")


alpha = 0.05
power = 0.80
mde_absolute = 0.01

z_alpha = stats.norm.ppf(1 - alpha / 2)
z_beta = stats.norm.ppf(power)
z_factor = (z_alpha + z_beta)**2

n_orig = (2 * var_original * z_factor) / (mde_absolute**2)
n_cupac = (2 * var_cupac * z_factor) / (mde_absolute**2)

print("\n--- ЭКОНОМИЯ ТРАФИКА (1 группа) ---")
print(f"Требуется ДО:    {int(n_orig):,} юзеров")
print(f"Требуется ПОСЛЕ: {int(n_cupac):,} юзеров")
print(f"💰 Экономия:      {int(n_orig - n_cupac):,} юзеров")
```

```python
import numpy as np
import pandas as pd
import scipy.stats as stats
from sklearn.ensemble import HistGradientBoostingRegressor
from sklearn.model_selection import KFold, cross_val_predict


# Шаг 7: Полный пайплайн CUPED для сегмента
segment_col = 'app_nm'
segment_val = 'mobile_ios'

# Копируем данные, чтобы не испортить основной датасет
df_segment = df_mb[df_mb[segment_col] == segment_val].copy()
target = 'life_1_day_flg'

print("=======================================================")
print(f"Анализируем СЕГМЕНТ: {segment_col} = '{segment_val}'")
print(f"Размер выборки сегмента: {df_segment.shape[0]:,} пользователей")
print("=======================================================\n")

# === 1. ЖЕСТКИЙ ОТБОР ЧИСТЫХ ФИЧЕЙ (Защита от Data Leakage) ===
bad_columns = [
    'life_1_day_flg', 'life_32_day_flg', 'orig_cost_sum',
    'nfu_flg', 'channel_desc', 'flow_strategy_code',
    'app_version_no', 'app_nm', 'full_apps'
]

numeric_cols = df_segment.select_dtypes(include=[np.number]).columns
features_seg = [
    col for col in numeric_cols
    if col not in bad_columns
    and 'cuped' not in col.lower()
    and 'cupac' not in col.lower()
    and 'pred' not in col.lower()
    and 'covariate' not in col.lower()
    and 'bin' not in col.lower()
]

# Считаем честные корреляции
correlations_seg = df_segment[features_seg].corrwith(df_segment[target])
top_covariates_seg = correlations_seg.abs().sort_values(ascending=False)

print(f"Топ-3 ЧЕСТНЫХ ковариат для сегмента {segment_val}:")
print("-" * 55)
for col in top_covariates_seg.head(3).index:
    print(f"{col:<30} | Корреляция: {correlations_seg[col]:.4f}")

# === 2. SINGLE-CUPED (Берем лучшую одиночную фичу сегмента) ===
best_single_feature = top_covariates_seg.index[0]
cov_single = df_segment[target].cov(df_segment[best_single_feature])
var_single_x = df_segment[best_single_feature].var()
theta_single = cov_single / var_single_x if var_single_x > 0 else 0
df_segment[f'{target}_cuped'] = df_segment[target] - theta_single * (df_segment[best_single_feature] - df_segment[best_single_feature].mean())

# === 3. ИТЕРАТИВНЫЙ MULTI-CUPED ===
y_iter = df_segment[target].copy()
used_features = []

for i in range(1000):
    corrs = df_segment[features_seg].corrwith(y_iter).abs().drop(labels=used_features, errors='ignore')
    if corrs.empty or corrs.max() < 0.005: break
    best_feat = corrs.idxmax()
    used_features.append(best_feat)

    cov_m = y_iter.cov(df_segment[best_feat])
    var_m = df_segment[best_feat].var()
    if var_m == 0: continue

    theta_m = cov_m / var_m
    y_iter = y_iter - theta_m * (df_segment[best_feat] - df_segment[best_feat].mean())

df_segment[f'{target}_multi_cuped'] = y_iter

# === 4. ЭТАЛОННЫЙ CUPAC (HistBoosting) ===
print("\n⏳ Обучаем CUPAC для сегмента...")
X_seg = df_segment[features_seg]
Y_seg = df_segment[target]

kf = KFold(n_splits=5, shuffle=True, random_state=42)
ml_model = HistGradientBoostingRegressor(max_iter=100, max_depth=4, random_state=42)
df_segment['ml_covariate_seg'] = cross_val_predict(ml_model, X_seg, Y_seg, cv=kf)

cov_cupac = df_segment[target].cov(df_segment['ml_covariate_seg'])
var_cupac_pred = df_segment['ml_covariate_seg'].var()
theta_cupac = cov_cupac / var_cupac_pred if var_cupac_pred > 0 else 0

df_segment[f'{target}_cupac'] = df_segment[target] - theta_cupac * (df_segment['ml_covariate_seg'] - df_segment['ml_covariate_seg'].mean())

# === 5. ОЦЕНКА РЕЗУЛЬТАТОВ ===
var_original = df_segment[target].var()
var_single = df_segment[f'{target}_cuped'].var()
var_multi = df_segment[f'{target}_multi_cuped'].var()
var_cupac = df_segment[f'{target}_cupac'].var()

print("\n--- БИТВА АЛГОРИТМОВ НА СЕГМЕНТЕ (Сжатие дисперсии) ---")
print(f"1. Single-CUPED ({best_single_feature}): {(1 - var_single/var_original)*100:.2f}%")
print(f"2. Iterative-CUPED (Топ {len(used_features)} фичей):     {(1 - var_multi/var_original)*100:.2f}%")
print(f"3. CUPAC (Gradient Boosting):         {(1 - var_cupac/var_original)*100:.2f}% 🔥")

# === 6. РАСЧЕТ ЭКОНОМИИ ТРАФИКА ДЛЯ СЕГМЕНТА ===
alpha = 0.05
power = 0.80
mde_absolute = 0.01

z_factor = (stats.norm.ppf(1 - alpha / 2) + stats.norm.ppf(power))**2
n_orig = (2 * var_original * z_factor) / (mde_absolute**2)
n_cupac_val = (2 * var_cupac * z_factor) / (mde_absolute**2)

print("\n--- ЭКОНОМИЯ ТРАФИКА В А/Б ТЕСТЕ (iOS сегмент) ---")
print(f"Требуется трафика ДО:    {int(n_orig):,} пользователей")
print(f"Требуется трафика ПОСЛЕ: {int(n_cupac_val):,} пользователей")
print(f"💰 Экономия (1 группа):   {int(n_orig - n_cupac_val):,} пользователей")
```
